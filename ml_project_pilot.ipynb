# **ML Project: Pilot**

# **Load Data**

```python
# y = f(x) : y is the value we need to predict
# logS column is the y variable we are predicting
# other columns are the x variables
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

# **Data Preparation**

## Data separation as X and y

```python
y = df["logS"]
y
```

```python
X = df.drop('logS', axis=1)
X
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
X_train
```

```python
X_test
```

# **Model Building**

## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

###**Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

###**Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("LR Train:", lr_train_mse, lr_train_r2)
print("LR Test:", lr_test_mse, lr_test_r2)
```

MSE on the training set and the test set are quite close, indicating that the model has similar performance
suggesting low variance and a good fit.

RÂ² score on the training set and the test set are also quite close, indicating that the model explains a
similar proportion of the variance

Consistency: The model performance is consistent across the training and test sets, which is a good sign as
it indicates that the model generalizes well to unseen data.
Performance: The RÂ² values suggest that the model has a reasonably good fit, explaining about
76% to 79% of the variance in the data

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

## **Random Forest**

### **Training the model**

```python
# if the y variable is a quantitative value - use regression model
# if the y variable is a categorical - use classification model
# in our case, y is quantitative, hence we're using a regression model

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

##**Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results])
df_models
```

```python
df_models.reset_index(drop=True)
```

For the training set, Linear Regression has a slightly lower MSE and a slightly higher RÂ² compared to Random Forest. This indicates that Linear Regression fits the training data marginally better than the Random Forest.

For the test set, Linear Regression again outperforms Random Forest, having a significantly lower MSE and higher RÂ². This suggests that Linear Regression not only fits the training data well but also generalizes better to unseen data.
